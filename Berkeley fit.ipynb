This notebook is trying to reproduce graphs on the page: http://berkeleyearth.org/results-summary/

We have downloaded the `annual-with-forcing-small.png` and `decadal-with-forcing-small.png` images and we are trying to reproduce them
below using the original data. The page says that these graphs were constructed using the `Full_TAVG_complete.txt` file. As we show below, this file does not seem to reproduce the graphs. However, the page also references an Excel file `http://berkeleyearth.org/xls/forcing-comparison.xlsx`, so we have saved the numbers for temperature, volcanic and CO2 concentrations from it into `best.data`. This file does seem to reproduce all the plots, see below. Although some confusion remains over what exact averaging was done in the original plots.

```python
%pylab inline
from IPython.display import Image
```

```python
D = loadtxt("data/best.data")
years = D[:, 0]
log_CO2 = D[:, 1]
volcanic = D[:, 2]
temp = D[:, 3]
unc = D[:, 4]

D = loadtxt("Full_TAVG_complete.txt", comments="%")
# Read from file:
T0 = 8.79
# Construct floating point years using years + months
years_int = D[:, 0]
months = D[:, 1]
years2 = years_int + (months-1) / 12
temp_month = D[:, 2] + T0
unc_month = D[:, 3]
temp1 = D[:, 4] + T0
unc1 = D[:, 5]
temp10 = D[:, 8] + T0
unc10 = D[:, 9]
```

Let's calculate the fit using the $\alpha$, $\beta$ and $\gamma$ values provides in the Excel file:

```python
alpha =  8.3421049
beta  =  4.4663687
gamma = -0.0151461
fit = alpha + beta * log_CO2 + gamma * volcanic
```

```python
def running_average(temp, n=6):
    """
    Calculates centered running average from -n to +n points (total of 2n+1 points).

    Example: if n = 6 and the data is monthly, then this function calculates the centered 13 month running average.
    """
    avg = temp.copy()
    for i in range(n, size(temp)-n-1):
        avg[i] = average(temp[i-n:i+n+1])
    avg[:n] = NaN
    avg[-n:] = NaN
    return avg
```

We plot the annual data from `best.data` (i.e. we plot it directly, no averaging is done). This seems to agree well with the original graph:

```python
figure(figsize=(8, 6))
fill_between(years, temp+unc, temp-unc, color="gray")
plot(years, temp, "k-")
plot(years, fit, "r-", lw=2)
xlabel("Year")
ylabel("Annual temperature [$^\circ C$]")
title("Berkeley Averaging method")
xlim([1750, 2013])
ylim([6.5, 10.5])
grid()
show()
Image(filename="data/annual-with-forcing-small.png")
```

For comparison, we plot the same graph, but using temperatures from the file `Full_TAVG_complete.txt`. This does not agree well:

```python
figure(figsize=(8, 6))
fill_between(years2, temp1+unc1, temp1-unc1, color="gray")
plot(years2, temp1, "k-")
plot(years, fit, "r-", lw=2)
xlabel("Year")
ylabel("Annual temperature [$^\circ C$]")
title("Berkeley Averaging method")
xlim([1750, 2013])
ylim([6.5, 10.5])
grid()
show()
Image(filename="data/annual-with-forcing-small.png")
```

Now we try to reproduce the other graph. We plot the 10 year average. Let's first use the data from `Full_TAVG_complete.txt`. As you can see, this does not agree well:

```python
figure(figsize=(8, 6))
fill_between(years2, temp10+unc10, temp10-unc10, color="gray")
plot(years2, temp10, "k-", label="Temperature (10 year average)")
plot(years, running_average(fit, 60), "r-", lw=2, label="fit (10 year moving average)")
xlabel("Year")
ylabel("Decadal temperature [$^\circ C$]")
title("Berkeley Averaging method")
xlim([1750, 2013])
ylim([7.5, 10])
grid()
legend(loc="lower right")
show()
Image(filename="data/decadal-with-forcing-small.png")
```

So we'll use the data from `best.data` and average it over a centered 10 year average. This seems to agree well:

```python
figure(figsize=(8, 6))
fill_between(years, running_average(temp+unc, 60), running_average(temp-unc, 60), color="gray")
plot(years, running_average(temp, 60), "k-")
plot(years, running_average(fit, 60), "r-", lw=2)
xlabel("Year")
ylabel("Decadal temperature [$^\circ C$]")
title("Berkeley Averaging method")
xlim([1750, 2013])
ylim([7.5, 10])
grid()
show()
Image(filename="data/decadal-with-forcing-small.png")
```

Unfortunately, assuming the `best.data` contains the annual temperatures, we actually did 121 year running average here. Not 10-year.
It would be 10 year if the `best.data` contained monthly data, but then our previous graph would disagree. So there is some confusion going on here.
