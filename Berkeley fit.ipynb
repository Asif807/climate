```python
D = loadtxt("data/best.data")
years = D[:, 0]
log_CO2 = D[:, 1]
volcanic = D[:, 2]
temp = D[:, 3]
unc = D[:, 4]

D = loadtxt("Full_TAVG_complete.txt", comments="%")
# Read from file:
T0 = 8.79
# Construct floating point years using years + months
years_int = D[:, 0]
months = D[:, 1]
years2 = years_int + (months-1) / 12
temp10 = D[:, 8] + T0
unc10 = D[:, 9]
```

```python
def running_average(temp):
    avg = temp.copy()
    n = 6*5
    for i in range(n, size(temp)-n-1):
        avg[i] = average(temp[i-n:i+n+1])
    avg[:n] = NaN
    avg[-n:] = NaN
    return avg
```

```python
alpha =  8.3421049
beta  =  4.4663687
gamma = -0.0151461
fit = alpha + beta * log_CO2 + gamma * volcanic

fill_between(years2, temp10+unc10, temp10-unc10, color="gray")
plot(years2, temp10, "k-")
plot(years, fit, "r-", lw=2)
xlabel("Year")
ylabel("Decadal temperature [$^\circ C$]")
title("Berkeley Averaging method")
xlim([1750, 2013])
ylim([7, 10])
grid()
show()
Image(filename="data/decadal.png")
```

```python
figure(figsize=(8, 6))
fill_between(years, temp+unc, temp-unc, color="gray")
plot(years, temp, "k-")
plot(years, fit, "r-", lw=2)
xlabel("Year")
ylabel("Decadal temperature [$^\circ C$]")
title("Berkeley Averaging method")
xlim([1750, 2013])
ylim([6.5, 10.5])
grid()
show()
Image(filename="data/annual-with-forcing-small.png")
```
