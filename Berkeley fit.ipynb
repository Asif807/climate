This notebook is trying to reproduce graphs on the page: http://berkeleyearth.org/results-summary/

We have downloaded the `annual-with-forcing-small.png` and `decadal-with-forcing-small.png` images and we are trying to reproduce them
below using the original data. The page says that these graphs were constructed using the `Full_TAVG_complete.txt` file. As we show below, this file does not seem to reproduce the graphs. However, the page also references an Excel file `http://berkeleyearth.org/xls/forcing-comparison.xlsx`, so we have saved the numbers for temperature, volcanic and CO2 concentrations from it into `best.data`. This file does seem to reproduce all the plots, see below. Although some confusion remains over what exact averaging was done in the original plots.

```python
%pylab inline
from IPython.display import Image
```

```python
D = loadtxt("data/best.data")
years = D[:, 0]
log_CO2 = D[:, 1]
volcanic = D[:, 2]
temp = D[:, 3]
unc = D[:, 4]

D = loadtxt("Full_TAVG_complete.txt", comments="%")
# Read from file:
T0 = 8.79
# Construct floating point years using years + months
years_int = D[:, 0]
months = D[:, 1]
years2 = years_int + (months-1) / 12
temp10 = D[:, 8] + T0
unc10 = D[:, 9]
```

```python
def running_average(temp, n=6):
    """
    Calculates centered running average from -n to +n points (total of 2n+1 points).

    Example: if n = 6 and the data is monthly, then this function calculates the centered 13 month running average.
    """
    avg = temp.copy()
    for i in range(n, size(temp)-n-1):
        avg[i] = average(temp[i-n:i+n+1])
    avg[:n] = NaN
    avg[-n:] = NaN
    return avg
```

We plot the 10 year average. For now we use the data from `Full_TAVG_complete.txt`, but this does not agree well.
So we must use the data from `best.data` instead and calculate the average ourselves.

```python
alpha =  8.3421049
beta  =  4.4663687
gamma = -0.0151461
fit = alpha + beta * log_CO2 + gamma * volcanic

figure(figsize=(8, 6))
fill_between(years2, temp10+unc10, temp10-unc10, color="gray")
plot(years2, temp10, "k-", label="Temperature (10 year average)")
plot(years, running_average(fit, 60), "r-", lw=2, label="fit (10 year moving average)")
xlabel("Year")
ylabel("Decadal temperature [$^\circ C$]")
title("Berkeley Averaging method")
xlim([1750, 2013])
ylim([7, 10])
grid()
legend(loc="lower right")
show()
Image(filename="data/decadal-with-forcing-small.png")
```

We plot the monthly data from `best.data`. This seems to agree the most with the original graph:

```python
figure(figsize=(8, 6))
fill_between(years, temp+unc, temp-unc, color="gray")
plot(years, temp, "k-")
plot(years, fit, "r-", lw=2)
xlabel("Year")
ylabel("Decadal temperature [$^\circ C$]")
title("Berkeley Averaging method")
xlim([1750, 2013])
ylim([6.5, 10.5])
grid()
show()
Image(filename="data/annual-with-forcing-small.png")
```

Let's average it over a centered 13 month average. This should be very close to a running 12 month average. Yet
the graphs are very different:

```python
figure(figsize=(8, 6))
#fill_between(years, temp+unc, temp-unc, color="gray")
plot(years, running_average(temp, 6), "k-")
plot(years, fit, "r-", lw=2)
xlabel("Year")
ylabel("Decadal temperature [$^\circ C$]")
title("Berkeley Averaging method")
xlim([1750, 2013])
ylim([6.5, 10.5])
grid()
show()
Image(filename="data/annual-with-forcing-small.png")
```

From all this, it seems that the file `best.data` contains the 12-month moving average of the temperatures.
